#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
calificaciones = [8.5, 9.2, 7.8, 8.9, 10.0]
promedio = sum(calificaciones) / len(calificaciones)
print(f'El promedio es: {promedio:.1f}')
```

- Encuentra la calificación más alta y la más baja.

```python
calif_baja = min(calificaciones)
calif_alta = max(calificaciones)
print(f'La calificacion mas baja es {calif_baja} y la mas alta es {calif_alta}')
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
ds["price"] = pd.to_numeric(ds["price"], errors="coerce")

filtro = (ds["level5"] == "Arroyomolinos (Madrid)") & (ds["price"] > 0)
media = ds.loc[filtro, "price"].mean()

print(f'La media es: {media:.2f}')
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
# Trazar el histograma de precios
import matplotlib.pyplot as plt

# Asegura que pricio sea numérico
ds["price"] = pd.to_numeric(ds["price"], errors="coerce")

# Filtrar población + precios válidos
filtro = (ds["level5"] == "Arroyomolinos (Madrid)") & (ds["price"] > 0)
precios = ds.loc[filtro, "price"]

# Histograma
plt.figure(figsize=(8,4))
plt.hist(precios, bins=20)
plt.title('Histograma de precios — Arroyomolinos (Madrid)')
plt.xlabel('Precio (€)')
plt.ylabel('Frecuencia')
plt.show()
```

Diria que en esta poblacion de Arroyomolinos se ve que la mayoría de las casas se mueven en precios normales o acorde al mercado inmobiliario de aca en España, sobre todo entre 200.000 y 390.000 €, y donde más se repite el precio es más o menos por la zona de 300.000 a 350.000 €. Luego hay algunas pocas que se van bastante arriba, más de 400.000 € y hasta cerca de 600.000 €, aunque diria que son casos muy puntuales. En resumen hay muchas casas en el rango medio y unas cuantas “premium” o "Top" que hacen que el gráfico tenga cola hacia la derecha.
